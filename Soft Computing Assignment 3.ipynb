```python
#Importing useful Libraries
import matplotlib.pyplot as plt
import numpy as np

#Global declaration on input x
x=np.linspace(-10,10,100)
```

```python
#Sigmoid Transfer Function
def sigmoid(x):
    z=1/(1+np.exp(-x))
    return z
```

```python
plt.plot(x,sigmoid(x))
plt.xlabel("x")
plt.ylabel("Sigmoid(x)")
plt.show
```

```python
#Binary Step Transfer Function
def binaryStep(x):
    return np.heaviside(x,1)
```

```python
plt.plot(x,binaryStep(x))
plt.xlabel("x")
plt.ylabel("Binary Step")
plt.show
```

```python
#Linear Transfer Function
def linear(x):
    return x
```

```python
plt.plot(x,linear(x))
plt.xlabel('x')
plt.ylabel('Linear')
plt.show
```

```python
def softmax(x):
    return ((np.exp(x))/np.sum(np.exp(x)))
```

```python
plt.plot(x,softmax(x))
plt.xlabel('x')
plt.ylabel('Softmax')
```

```python
#Tanh Transfer Function
def tanh(x):
    return (1-np.exp(-2*x))/(1+np.exp(-2*x))
```

```python
plt.plot(x,tanh(x))
plt.xlabel('x')
plt.ylabel('Tanh')
plt.show()
```

```python
#Rectified Linear unit transfer function(ReLu)
def ReLu(x):
    input=[]
    for i in x:
        if i>0:
            input.append(i)
        else:
            input.append(0)
    return input
            
```

```python
plt.plot(x,ReLu(x))
plt.xlabel('x')
plt.ylabel('ReLu')
```

Q2. Provied detailed analysis of RELU transfer function along with some use case?

In a neural network, the transfer function is responsible for transforming the summed weighted input into the activation of 
the node.
Rectified Linear Unit (ReLU) function is improved version of linear function where linear function returns its input value for all input x whereas ReLu only returns its input for all x greater than 0. ReLu deactivates the all neurons present in the negative region.

Purpose of ReLU

Traditionally, some prevalent non-linear activation functions, like sigmoid functions (or logistic) and hyperbolic tangent,
are used in neural networks to get activation values corresponding to each neuron. Recently, the ReLu function has been used 
instead to calculate the activation values in traditional neural network or deep neural network paradigms. The reasons of 
replacing sigmoid and hyperbolic tangent with ReLu consist of:

Computation saving - the ReLU function is able to accelerate the training speed of deep neural networks compared to traditional
activation functions since the derivative of ReLU is 1 for a positive input. Due to a constant, deep neural networks do not 
need to take additional time for computing error terms during training phase.

Solving the vanishing gradient problem
The ReLU function does not trigger the vanishing gradient problem when the number of layers grows. This is because this functiondoes not have an asymptotic upper and lower bound. Thus, the earliest layer (the first hidden layer) is able to receive the errors coming from the last layers to adjust all weights between layers. By contrast, a traditional activation function like sigmoid is restricted between 0 and 1, so the errors become small for the first hidden layer. This scenario will lead to a 
poorly trained neural network.

Advantages for ReLU Function:
1. Unlike Tanh and Sigmoid function, ReLu make computation simpler.
2. Most of the time ReLu function shows linear linear behavior and in general a neural netwok is easier to optimize which has 
   linear or close to linear behavior.
3. The most important feature of ReLU function it give output value exact 0 whereas Tanh and Sigmoid function has approximate 
   0 output which does not represent close relation.
4. Another advantage of ReLU function it reduces the vanishing of gradient.
